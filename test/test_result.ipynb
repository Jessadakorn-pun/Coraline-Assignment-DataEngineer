```python
import os
import pandas as pd
from sqlalchemy import create_engine
```

## Create Connection String and Connect to Data Warehose

```python
user = os.getenv("DATA_WAREHOUSE_POSTGRES_USER")
password = os.getenv("DATA_WAREHOUSE_POSTGRES_PASSWORD")
port = os.getenv("DATA_WAREHOUSE_POSTGRES_PORT")
db = os.getenv("DATA_WAREHOUSE_POSTGRES_DB")

engine = create_engine(
    f"postgresql+psycopg2://{user}:{password}@localhost:{port}/{db}"
)
```

## Result

### food_sales Table

```python
query = """

SELECT *
FROM food_sales
LIMIT 10

"""

df = pd.read_sql(query, engine)
df
```

### cate_reg Table

```python
query = """

SELECT *
FROM cat_reg

"""

df = pd.read_sql(query, engine)
df
```
